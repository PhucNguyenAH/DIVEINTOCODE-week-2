## Problem 5

```python
def compute_gradient(function, x_range=(-50, 50, 0.1)):
    """
    Calculate the gradient using the amount of change.
    Parameters
    ----------------
    function : function
      The function you want to find, the one that returns the ndarray of y when you put the ndarray of x.
    x_range : tuple
      Specify the range in the same way as the argument of np.arange ().
    Returns
    ----------------
    array_xy : ndarray, shape(n, 2)
      A combination of x and y. n depends on x_range.
    gradient : ndarray, shape(n-1,)
      Function gradient. One factor is reduced to make a difference
    """
    start, stop, step = x_range
    X = np.arange(start, stop, step)
    y = function(X)
    array_xy = np.vstack((X,y)).T
    d_X = X[1:]-X[:-1]
    d_y = y[1:]-y[:-1]
    gradient = d_y/d_X
    last_slope = gradient[-1]
    gradient = np.append(gradient, last_slope) 
    plt.title("Draw the function")
    plt.xlabel("X")
    plt.ylabel("gradient")
    plt.plot(X, gradient) 
    plt.show()
    return array_xy, gradient
```

```python
def function1(array_x):
    array_y = array_x**2
    return array_y
```

```python
def function2(array_x):
    array_y = 2*(array_x**2) + 2**array_x
    return array_y
```

```python
def function3(array_x):
    array_y = np.sin(array_x**(1/2))
    return array_y
```

```python
array_xy, gradient = compute_gradient(function1)
print("Combine the ndarrays created in Problem 1: \n -Shape: {} \n -Values: \n {}".format(array_xy.shape,array_xy))
print("The slope at each x coordinate: {}".format(gradient))
```

```python
array_xy, gradient = compute_gradient(function2)
print("Combine the ndarrays created in Problem 1: \n -Shape: {} \n -Values: \n {}".format(array_xy.shape,array_xy))
print("The slope at each x coordinate: {}".format(gradient))
```

```python
array_xy, gradient = compute_gradient(function3, (0,50,0.1))
print("Combine the ndarrays created in Problem 1: \n -Shape: {} \n -Values: \n {}".format(array_xy.shape,array_xy))
print("The slope at each x coordinate: {}".format(gradient))
```

## Problem 6

```python
X = np.arange(-50, 50, 0.1)
y = function1(X)
y_min = y.min()
y_argmin = y.argmin()
print("Function y = x^2 (-50<=x<=50): \n The minimum value of y: {} at index {}".format(y_min, y_argmin))
print("Display the slope before taking the minimum y")
compute_gradient(function1)
print("Display the slope after taking the minimum y")
X_after = np.delete(X,y_argmin)
y_after = np.delete(y,y_argmin)
d_X = X_after[1:]-X_after[:-1]
d_y = y_after[1:]-y_after[:-1]
gradient = d_y/d_X
last_slope = gradient[-1]
gradient = np.append(gradient, last_slope) 
plt.title("Draw the function")
plt.xlabel("X")
plt.ylabel("gradient")
plt.plot(X_after, gradient) 
plt.show()
```
